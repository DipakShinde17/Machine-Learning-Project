```python
data = pd.read_csv("ai_resume_screening.csv")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
def showPlot():    
    plt.subplot(1, 2, 1)
    plt.boxplot(data["years_experience"])
    plt.subplot(1, 2, 2)
    sns.histplot(data = data, x = data["years_experience"], kde=True)
    plt.axvline(data["years_experience"].mean()+3*data["years_experience"].std(), color='red')
    plt.axvline(data["years_experience"].mean()-3*data["years_experience"].std(), color='red')
showPlot()
```
