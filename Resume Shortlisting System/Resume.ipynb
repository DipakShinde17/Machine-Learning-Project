### Fill Outlier

```python
Q1 = data["skills_match_score"].quantile(0.25)
Q3 = data["skills_match_score"].quantile(0.75)

IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

lower_limit, upper_limit
```

```python
data["skills_match_score"] = data["skills_match_score"].clip(lower=lower_limit, upper=upper_limit)
```

```python
def showPlot():    
    plt.subplot(1, 2, 1)
    plt.boxplot(data["skills_match_score"])
    plt.subplot(1, 2, 2)
    sns.histplot(data = data, x = data["skills_match_score"], kde=True)
    plt.axvline(data["skills_match_score"].mean()+3*data["skills_match_score"].std(), color='red')
    plt.axvline(data["skills_match_score"].mean()-3*data["skills_match_score"].std(), color='red')
showPlot()
```

### Check Correlation

```python
sns.heatmap(data.corr(),annot=True)
```

```python
x = data.drop('shortlisted', axis=1)
y = data['shortlisted']
```

### Build model

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=42)
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(x_train)
X_test = scaler.transform(x_test)
```

## Imbalanced Data

```python
data["shortlisted"].value_counts()
```

## Handle Imbalance Data

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(random_state=42)

X_train_sm, y_train_sm = smote.fit_resample(X_train, y_train)
```

```python
print("Before SMOTE:")
print(y_train.value_counts())
```

```python
print("\nAfter SMOTE:")
print(y_train_sm.value_counts())
```

# LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter=1000,class_weight='balanced') 
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
prob = model.predict_proba(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
print("testing Accuracy:", accuracy_score(y_test, y_pred))
```

```python
train_pred = model.predict(X_train)
print("Training Accuracy:", accuracy_score(y_train, train_pred))
```

```python
print(confusion_matrix(y_test, y_pred))               #It shows how many predictions are correct and how many are incorrect.
```

```python
print(classification_report(y_test, y_pred))
```

```python
 # precision :  Correct positive prediction
# recall : detecting all positive
# f1 score : balance of precision & recall
# support : number of actual occurences
```

### Cross Validate

```python
from sklearn.model_selection import cross_validate
```

```python
knn = LogisticRegression(max_iter=1000)
val_result = cross_validate(knn, x_train, y_train,cv=9)
val_result
```

```python
train_acc = val_result['test_score'].mean()
print("Training acc:", train_acc)
```
